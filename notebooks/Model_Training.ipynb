```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_Cleaned.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
## Droping day month year
df.drop(['day', 'month', 'year'],axis=1,inplace=True)
```

```python
df.head()
```

```python
## Encoding Classes

df['Classes'] = np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.head()
```

```python
df.tail()
```

```python
## Independent and Dependent Features

X = df.drop('FWI',axis=1)
y = df['FWI']
```

```python
X.head()
```

```python
y.head()
```

```python
## Trai Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train.shape,X_test.shape
```

```python
## Feature Selection Based on Correlation
X_train.corr()
```

```python
## Check for Multicolinearity
plt.figure(figsize=(16,14))
corr = X_train.corr()
sns.heatmap(corr,annot=True)
```

```python
## A Function for Removing a Feature Based on Threshold
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## The Threshold is decieded by the Domain Expert
corr_features = correlation(X_train,0.85)
```

```python
## Drop the Features When Correlation is more than the Threshold that is 0.85

X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
X_test.shape,X_train.shape
```

# Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
x_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
x_test_scaled
```

# Box Plots To understand Effect Of Standard Scaler

```python
plt.subplots(figsize = (20,10))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error , r2_score
regression = LinearRegression()
regression.fit(X_train_scaled,y_train)
y_pred = regression.predict(x_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test , y_pred)
print(f'Mean Absolute Error:{mae}\nR2 Score:{score}')
plt.scatter(y_test,y_pred)
```

# Lasso Regression


```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error , r2_score
L = Lasso()
L.fit(X_train_scaled,y_train)
y_pred = L.predict(x_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test , y_pred)
print(f'Mean Absolute Error:{mae}\nR2 Score:{score}')
plt.scatter(y_test,y_pred)
```

# Ridge Regression Model


```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error , r2_score
R = Ridge()
R.fit(X_train_scaled,y_train)
y_pred = R.predict(x_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test , y_pred)
print(f'Mean Absolute Error:{mae}\nR2 Score:{score}')
plt.scatter(y_test,y_pred)
```

# Elasticnet Regression


```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error , r2_score
EN = ElasticNet()
EN.fit(X_train_scaled,y_train)
y_pred = EN.predict(x_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test , y_pred)
print(f'Mean Absolute Error:{mae}\nR2 Score:{score}')
plt.scatter(y_test,y_pred)
```

# Cross Validation Lasso

```python
from sklearn.linear_model import LassoCV
from sklearn.metrics import mean_absolute_error , r2_score
LCV = LassoCV(cv=5)
LCV.fit(X_train_scaled,y_train)
y_pred = LCV.predict(x_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
score = r2_score(y_test , y_pred)
print(f'Mean Absolute Error:{mae}\nR2 Score:{score}')
plt.scatter(y_test,y_pred)
```
